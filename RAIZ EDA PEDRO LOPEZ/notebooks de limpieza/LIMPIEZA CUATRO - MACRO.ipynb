```python
df['2023'] = df['2023'].round(2)
```

```python
df['2023'] = df['2023'].astype(str)  # convertir a texto
df['2023'] = df['2023'].str.replace('.', '', regex=False)  # quita separador de miles
df['2023'] = df['2023'].str.replace(',', '.', regex=False)  # coma a punto decimal
df['2023'] = pd.to_numeric(df['2023'], errors='coerce')
```

```python
df
```

```python
df['grupo_PIB'] = pd.qcut(df['2023'], q=3, labels=['Bajo', 'Medio', 'Alto'])
```

```python
df = df.drop(['2024','Unnamed: 69'], axis =1)
```
